```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
import threading

lock = threading.Lock()

class Response:
    def __str__(self):
        return "You called the Class Object!"

    def getResponse(self, url):
        self.link = url
        soup = BeautifulSoup(self.link.text, "html.parser")

        lock.acquire()
        
        ratings = []
        # Getting ratings of top action movies
        divTags = soup.find_all('div', class_ = 'col-imdb-rating')
        for tag in divTags:
            if tag.find('strong') is not None:
                ratings.append(tag.find('strong').text.strip())
        
        movies = []
        # Getting corresponding movie names
        spanTags = soup.find_all('span', class_ = 'lister-item-header')
        for tag in spanTags:
            if tag.find('a') is not None:
                movies.append(tag.find('a').text.strip())

        dataSet = list(zip(ratings,movies))
        lock.release()
        return dataSet
        # print (ratings)
        # print()
        # print (movies)

class Thread1(threading.Thread):
    def __init__(self, resp):
        threading.Thread.__init__(self)
        self.resp = resp

    def __str__(self):
        return "You called the thread object!"

    def run(self):
        time.sleep(1)
        url = requests.get("https://www.imdb.com/search/title?genres=action&sort=user_rating,desc&title_type=feature&num_votes=25000,&pf_rd_m=A2FGELUUNOQJNL&pf_rd_p=5aab685f-35eb-40f3-95f7-c53f09d542c3&pf_rd_r=NBMH7HZ8F7T2JJK17T31&pf_rd_s=right-6&pf_rd_t=15506&pf_rd_i=top&view=simple&ref_=chttp_gnr_1")
        data = self.resp.getResponse(url)
        df = pd.DataFrame(data = data, columns = ['Ratings', 'Names'])
        df.to_csv('Category1.csv', index = False, header = None)
        print('Data written successfully to Category1!')

class Thread2(threading.Thread):
    def __init__(self, resp):
        threading.Thread.__init__(self)
        self.resp = resp

    def __str__(self):
        return "You called the thread object!"

    def run(self):
        time.sleep(1)
        url = requests.get("https://www.imdb.com/search/title?genres=comedy&sort=user_rating,desc&title_type=feature&num_votes=25000,&pf_rd_m=A2FGELUUNOQJNL&pf_rd_p=5aab685f-35eb-40f3-95f7-c53f09d542c3&pf_rd_r=NBMH7HZ8F7T2JJK17T31&pf_rd_s=right-6&pf_rd_t=15506&pf_rd_i=top&view=simple&ref_=chttp_gnr_5")
        data = self.resp.getResponse(url)
        df = pd.DataFrame(data = data, columns = ['Ratings', 'Names'])
        df.to_csv('Category2.csv', index = False, header = None)
        print('Data written successfully to Category2!')

class Thread3(threading.Thread):
    def __init__(self, resp):
        threading.Thread.__init__(self)
        self.resp = resp

    def __str__(self):
        return "You called the thread object!"

    def run(self):
        time.sleep(1)
        url = requests.get("https://www.imdb.com/search/title?genres=romance&sort=user_rating,desc&title_type=feature&num_votes=25000,&pf_rd_m=A2FGELUUNOQJNL&pf_rd_p=5aab685f-35eb-40f3-95f7-c53f09d542c3&pf_rd_r=NBMH7HZ8F7T2JJK17T31&pf_rd_s=right-6&pf_rd_t=15506&pf_rd_i=top&view=simple&ref_=chttp_gnr_16")
        data = self.resp.getResponse(url)
        df = pd.DataFrame(data = data, columns = ['Ratings', 'Names'])
        df.to_csv('Category3.csv', index = False, header = None)
        print('Data written successfully to Category3!')

class Thread4(threading.Thread):
    def __init__(self, resp):
        threading.Thread.__init__(self)
        self.resp = resp

    def __str__(self):
        return "You called the thread object!"

    def run(self):
        time.sleep(1)
        url = requests.get("https://www.imdb.com/search/title?genres=history&sort=user_rating,desc&title_type=feature&num_votes=25000,&pf_rd_m=A2FGELUUNOQJNL&pf_rd_p=5aab685f-35eb-40f3-95f7-c53f09d542c3&pf_rd_r=NBMH7HZ8F7T2JJK17T31&pf_rd_s=right-6&pf_rd_t=15506&pf_rd_i=top&view=simple&ref_=chttp_gnr_11")
        data = self.resp.getResponse(url)
        df = pd.DataFrame(data = data, columns = ['Ratings', 'Names'])
        df.to_csv('Category4.csv', index = False, header = None)
        print('Data written successfully to Category4!')

class Thread5(threading.Thread):
    def __init__(self, resp):
        threading.Thread.__init__(self)
        self.resp = resp

    def __str__(self):
        return "You called the thread object!"

    def run(self):
        time.sleep(1)
        url = requests.get("https://www.imdb.com/search/title?genres=sci_fi&sort=user_rating,desc&title_type=feature&num_votes=25000,&pf_rd_m=A2FGELUUNOQJNL&pf_rd_p=5aab685f-35eb-40f3-95f7-c53f09d542c3&pf_rd_r=NBMH7HZ8F7T2JJK17T31&pf_rd_s=right-6&pf_rd_t=15506&pf_rd_i=top&view=simple&ref_=chttp_gnr_17")
        data = self.resp.getResponse(url)
        df = pd.DataFrame(data = data, columns = ['Ratings', 'Names'])
        df.to_csv('Category5.csv', index = False, header = None)
        print('Data written successfully to Category5!')


r1 = Response()
t1 = Thread1(r1)
t2 = Thread2(r1)
t3 = Thread3(r1)
t4 = Thread4(r1)
t5 = Thread5(r1)
t1.start()
t2.start()
t3.start()
t4.start()
t5.start()
```
